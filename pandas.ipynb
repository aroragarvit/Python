# Selecting rows using iloc and loc 
loc is used when selecting throuch index values and iloc is used when selecting through indexing number 

```python
import pandas as pd 
wine_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';')
wine_dfcopy=wine_df.copy()
wine_df.head()
print("shape of wine_dfcopy")
print(wine_df.shape)
```

```python
wine_df.nunique()
```

```python
wine_df = wine_df.drop_duplicates(subset='density')
print("shape of wine_df")
wine_df.shape #436 unique values in the density column 

```

#  setting a column as  a index 

```python
wine_df.set_index('density',inplace=True)
                 
```

```python
wine_df.head()
```

```python
(wine_df.iloc[[2]])
(wine_df.loc[[0.9970]])
```

```python
wine_dfcopy.iloc[[1, 4, 7]]   
```

```python
wine_df.iloc[[1,4,7]]  # in these some of the rows are dropped from between thats why wine_dfcopy and wine_df are giving diff values 
```

```python
wine_df.loc[[0.9968,0.9964,0.9943]]
```
